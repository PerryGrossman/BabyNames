```python
# This function gives us information about the boy name with the highest proportion in any given year.
# https://docs.python.org/2/tutorial/controlflow.html#defining-functions
# The keyword def introduces a function definition. 
#It must be followed by the function name and the parenthesized list of formal parameters. 
#The statements that form the body of the function start at the next line, and must be indented.
def get_max_record(group):
    return group.ix[group.prop.idxmax()]
get_max_record(boys)                   
```

```python
# Find the 50% percentile for names in 1880.
df = boys[boys.year == 1880].sort_index(by='prop',ascending=False)
df.prop.values.cumsum().searchsorted(0.5)
```

The 50% percentile is reached by the fifteenth name in 1880. In other words, one-half of all the babies born in 1880 had only one of 15 names.

```python
df = boys[boys.year == 2005].sort_index(by='prop',ascending=False)
# how many to get to 50%?
df.prop.values.cumsum().searchsorted(0.5)
```

The 50% percentile is reached by the 110th name in 2005. In other words, one-half of all the babies born in 2005 had one of 110 names.

```python
# For the whole dataset (from 1880 to 2008) we see that the 50% percentile is reached at record 127.

df.prop.cumsum()[:130]
#first 130 
# measure of diversity
```

The names data frame gives the proportion, but not the total number of people.



```python
# So let's bring in the numbers file
births = read_csv('births.csv')
```

```python
# pandas join function is merge
merge(names, births, on= [ 'year', 'sex'])

#simple join on year and sex in both Df.
# do three different examples of this.
```

```python
merged = merge(names, births, on= [ 'year', 'sex'])
```

```python
# total number,
# need to add the number of persons
merged['persons'] = np.floor(merged.prop * merged.births)

#decimal numbers so use npl.floor
```

```python
merged.head()
```

```python
merged.groupby(['name','sex'])['persons'].sum()
#for each name sex combo-- the total number of births.
#hierarchical labeling.
```

```python
# order for most popular names.
#df[::-1] reverses the order of the DataFrame's columns.
namesordered = merged.groupby(['name','sex'])['persons'].sum().order()
namesordered[::-1]
#for each name sex combo-- the total number of births.
#hierarchical labeling.
```
